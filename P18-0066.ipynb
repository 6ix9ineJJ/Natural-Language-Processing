```python
import spacy
import urduhack
import random
```

```python
with open("ghalib.txt","r") as f:
    corpus = f.read()
```

```python
import spacy 
nlp = spacy.blank('ur')
```

```python
doc = nlp(corpus)

#creating list of unigrams and bigrams
unigrams = [token.text for token in doc if not token.is_punct and not token.is_space]
bigrams = [(doc[i].text, doc[i+1].text) for i in range(len(doc)-1) if not doc[i].is_punct and not doc[i+1].is_punct and not doc[i].is_space and not doc[i+1].is_space]
```

```python
def generate_unigram():
    verse = ""
    for i in range(random.randint(7,10)):
        word = random.choice(unigrams)
        verse +=word + " "
    return verse.strip()

def generate_bigram():
    verse = ""
    last_word = ""
    for i in range(random.randint(7,10)):
        if i==0:
            word = random.choice(unigrams)
        else:
            possible_words = [b[1] for b in bigrams if b[0] == last_word]
            if possible_words:
                word = random.choice(possible_words)
            else:
                word = random.choice(unigrams)
        verse +=word + " "
        last_word = word
    return verse.strip()
```

```python
for i in range(3):
    for j in range(4):
        if j %2 == 0:
            verse = generate_unigram()
        else:
            verse = generate_bigram()
            print(verse)
        print()
```

```python
with open("faiz.txt","r") as f:
    corpus1 = f.read()
```

```python
from collections import Counter #count the frequency of items in list 
```

```python
doc = nlp(corpus1)
print(corpus1)
```

```python
words = [token.text for token in doc if not token.is_punct and not token.is_space]
print(words)
```

```python
#generate unigram  and bigrams models
unigrams = Counter(words)
bigrams = Counter(zip(words, words[1:]))
```

```python
def generate_verse():
    length = random.randint(7, 10)
    verse = [random.choice(words)]
    
    while len(verse) < length:    
        # Get the last word in the verse
        last_word = verse[-1]
        choices = [word for word in bigrams if word[0] == last_word]
        if not choices:
            return generate_verse()
        next_word = random.choice(choices)[1]
        verse.append(next_word)
    return ' '.join(verse)
    #generate poem
    for i in range(3):
        for j in range(4):
            verse = generate_verse()
            print(verse)
        print()
```
